```python
import pandas as pd
import numpy as np

# Given census data
census_data = pd.DataFrame({
    "State": ["Maharashtra", "Maharashtra", "Karnataka", "Karnataka", "Uttar Pradesh", "Uttar Pradesh", "Delhi", "Rajasthan", "Rajasthan", "Tamil Nadu", "Tamil Nadu"],
    "District": ["Mumbai", "Pune", "Bengaluru", "Mysuru", "Lucknow", "Kanpur", "Delhi", "Jaipur", "Udaipur", "Chennai", "Coimbatore"],
    "Population": [12400000, 3500000, 8500000, 1000000, 3000000, 2000000, 22000000, 3500000, 1500000, 8500000, 2000000],
    "Median_Age": [28.5, 29.2, 27.8, 26.5, 25.8, 26.2, 29.5, 27.2, 26.8, 28.2, 27.5],
    "Median_Household_Income": [350000, 280000, 420000, 220000, 180000, 200000, 500000, 250000, 200000, 380000, 280000],
    "Literacy_Rate": [85.2, 82.5, 88.5, 80.2, 75.1, 78.5, 91.2, 80.5, 78.2, 86.5, 84.2],
    "Urban_Population": [90.0, 70.0, 95.0, 60.0, 50.0, 55.0, 100.0, 65.0, 50.0, 90.0, 70.0]
})

# Calculate how many additional rows are needed
additional_rows = 220 - census_data.shape[0]

# Define cities for each state
cities = {
    "Maharashtra": ["Mumbai", "Pune", "Nagpur", "Aurangabad",'Ahmednagar','satara','Sangli','Latur','Kolhapur'],
    "Karnataka": ["Bengaluru", "Mysuru", "Mangaluru", "Hubli",'Raichur','Kolar','Bagalkot','Shimoga','Tumkur'],
    "Uttar Pradesh": ["Lucknow", "Kanpur", "Agra", "Varanasi",'Ayodhya','Noida','Saharanpur','Bareilly','Prayagraj'],
    "Delhi": ["South Delhi",'Central Delhi','	East Delhi','North Delhi','Shahdara','South East Delhi'],
    "Rajasthan": ["Jaipur", "Udaipur", "Jodhpur", "Bikaner",'Beawar','Kekri','Tonk','Ajmer','Nagaur'],
    "Tamil Nadu": ["Chennai", "Coimbatore", "Madurai", "Tiruchirappalli",'Kanniyakumari','Perambalur','Ranipet','Tenkasi','Viluppuram']
}

# Generate new unique data for additional rows
new_states = np.random.choice(list(cities.keys()), size=additional_rows)
new_districts = [np.random.choice(cities[state]) for state in new_states]
new_population = np.random.randint(1000000, 25000000, size=additional_rows)
new_median_age = np.random.uniform(20.0, 40.0, size=additional_rows)  # Broader range for median age
new_median_household_income = np.random.randint(150000, 1000000, size=additional_rows)  # Broader range for income
new_literacy_rate = np.random.uniform(60.0, 100.0, size=additional_rows)  # Broader range for literacy rate
new_urban_population = np.random.uniform(0.0, 100.0, size=additional_rows)  # Urban population percentage

# Create a new DataFrame with the unique new data
new_data = pd.DataFrame({
    "State": new_states,
    "District": new_districts,
    "Population": new_population,
    "Median_Age": new_median_age,
    "Median_Household_Income": new_median_household_income,
    "Literacy_Rate": new_literacy_rate,
    "Urban_Population": new_urban_population
})

# Append the new unique data to the existing data
census_data = pd.concat([census_data, new_data], ignore_index=True)

census_data
```
